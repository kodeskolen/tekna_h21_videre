## Sagtannbølge

Sagtannbølgen har fått navnet sitt fordi grafen ser ut som taggene på en sag. Sjekk ut i grafen under! I dette eksempelet bruker vi tonen _liten g_ med frekvens 330Hz..

Forresten, pass på ørene når du spiller av lyden!

```python
# Evaluer celle - Sagtannbølge
frekvens = 330
varighet = 2.
t = linspace(0., varighet, int(varighet * rate))
sagtannbølge = signal.sawtooth(frekvens * pi * 2 * t)  # signal.sawtooth: fra scipy

spill_av(sagtannbølge)



# ------------------------------------------------------------
# Kode for å plotte grafen:
x = arange(0, 900)  # Viser de 900 første målepunktene
figure().set_figwidth(15)
title(f"Sagtannbølge {frekvens} Hz")
xlabel("samplenummer (tid)")
ylabel("amplitude (lydstyrke)")
axline((0,0), (1,0), linewidth=0.75, color='r', linestyle=":")
plot(x,sagtannbølge[0:900])
show()
```

Vi skal også lage en `sagtann_generator` som lager sagtannbølger for oss:

```python
# Evaluer

def sagtann_generator(frekvens=440, varighet=1.0, amplitude=1.0):
    t = linspace(0., varighet, int(varighet * rate))
    sagtannbølge = signal.sawtooth(frekvens * pi * 2 * t) * amplitude
    return sagtannbølge
```

Sammenlign hvordan de forskjellige bølgeformene høres ut. Hører du forskjell på dem?

```python
# Evaluer først denne
sagtannbølge = sagtann_generator(330, 1)
spill_av(sagtannbølge)
```

```python
# så denne
sinusbølge = sinus_generator(330, 1)
spill_av(sinusbølge)
```

## Trekantbølge

Trekantbølgen har navnet sitt naturlig nok fra at den ser ut som en trekant. Når vi gir `signal.sawtooth` 0.5 som andre argument returnerer funksjonen en trekantbølge. Dette står det mer om i dokumentasjonen til Scipy.

Spill av trekantbølgen under.

```python
# Evaluer celle - trekantbølge
frekvens = 350
varighet = 2.
t = linspace(0., varighet, int(varighet * rate))
trekantbølge = signal.sawtooth(frekvens * pi * 2 * t, 0.5)  # signal.sawtooth: fra scipy

spill_av(trekantbølge)



# ------------------------------------------------------------
# Kode for å plotte grafen:b
x = arange(0, 900)  # Viser de 900 første målepunktene
figure().set_figwidth(15)
title(f"Trekantbølge {frekvens} Hz")
xlabel("samplenummer (tid)")
ylabel("amplitude (lydstyrke)")
axline((0,0), (1,0), linewidth=0.75, color='r', linestyle=":")
plot(x,trekantbølge[0:900])
show()
```

La oss definere en funksjon som gjør det enklere å lage trekantbølger:

```python
# Evaluer

def trekant_generator(frekvens=440, varighet=1.0, amplitude=1.0):
    t = linspace(0., varighet, int(varighet * rate))
    sagtannbølge = signal.sawtooth(frekvens * pi * 2 * t, 0.5) * amplitude
    return sagtannbølge
```

Sammenligne nå med sinusbølgen. Hører du at de låter forskjellig?

```python
# Evaluer først denne
trekantbølge = trekant_generator(350, 1)
spill_av(trekantbølge)
```

```python
# så denne
sagtannbølge = sagtann_generator(350, 1)
spill_av(sagtannbølge)
```

```python
# og så denne
sinusbølge = sinus_generator(350, 1)
spill_av(sinusbølge)
```
