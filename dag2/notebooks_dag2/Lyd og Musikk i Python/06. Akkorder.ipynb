# Akkorder

På instrumenter som piano eller gitar spiller man ofte flere toner samtidig. En slik kombinasjon av toner kalles en *akkord*. Man kan lage akkorder som både konsonerer, dissonerer, eller noe midt i mellom, avhengig av om tonene i akkorden "passer godt" til hverandre.

```python
# Evaluer først denne

from startscript import *

%matplotlib inline
```

For å finne ut hvilke frekvens en tone har bruker vi denne tabellen: https://no.wikipedia.org/wiki/Tangentfrekvenser_(piano). Her ser vi at kammertonen _enstrøken A_ (tangentnummer 40) har frekvens 440, slik vi så tidligere. Vi ser også at tonen c har frekvens ca. 260Hz.

Vi kan nå lage akkorder ved å summere signalene fra flere sinus-generatorer. Spill av de fire neste kode-cellene (akkordene) etter hverandre. Kanskje har du hørt dem før i noen låter?

```python
varighet = 2.  # Bruke samme varighet til alle signalene

c = sinus_generator(260, varighet) # Tonen c,
e = sinus_generator(330, varighet) # e,
g = sinus_generator(390, varighet) # og g
akkord = c + e + g                 # blir tilsammen akkorden c-dur

spill_av(akkord)
```

```python
varighet = 2.

a = sinus_generator(220, varighet) # Tonen a,
c = sinus_generator(260, varighet) # c
e = sinus_generator(330, varighet) # og e
akkord = a + c + e                 # blir akkorden a-moll

spill_av(akkord)
```

```python
varighet = 2.

f = sinus_generator(350, varighet) # Tonen f
a = sinus_generator(220, varighet) # a,
c = sinus_generator(260, varighet) # og c
akkord = f + a + c                 # blir akkorden f-dur

spill_av(akkord)
```

```python
varighet = 2.

g = sinus_generator(390, varighet) # Tonen g
b = sinus_generator(245, varighet) # b (også kalt h),
d = sinus_generator(290, varighet) # og d,
akkord = g + b + d                 # blir akkorden g-dur

spill_av(akkord)
```

## Grafen til en akkord

Evaluer cellen under for å se hvordan lydsignalet til akkorden C-dur ser ut. Signalet er altså summen av tre sinustoner.

```python
varighet = 2.

c = sinus_generator(260, varighet) # Tonen c,
e = sinus_generator(330, varighet) # e,
g = sinus_generator(390, varighet) # og g
akkord = c + e + g                 # blir akkorden c-dur

spill_av(akkord)


# ----------------------------------------
# Plotter akkord-lydsignalet som består av tre ulike toner, i tillegg til enkelttonene
alle_signaler = [c, e, g]
antall_samples = 1200  # Antall samples som skal vises
fig, axs = subplots(2)
fig.set_figwidth(15)
fig.set_figheight(7)
x = arange(0, antall_samples)
axs[0].set_title("Akkord")
axs[1].set_title("Enkelttoner")
axs[0].plot(x, akkord[:antall_samples])
for sig in alle_signaler:
    axs[1].plot(x, sig[:antall_samples])
show()
```
