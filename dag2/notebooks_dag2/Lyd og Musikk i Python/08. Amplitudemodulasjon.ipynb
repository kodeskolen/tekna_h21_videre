# Amplitudemodulasjon - AM

Tidligere har vi summert flere signaler, for eksempel sinusbølger, slik at de kombineres og vi hører flere lyder samtidig. Men går det an å multiplisere sinusbølger med hverandre?

```python
# Evaluer først denne

from startscript import *

%matplotlib inline
```

Evaluer cellen under, hør resulatet og ta en kikk på grafen.

```python
v = 3  # varighet
frekvens = 800
modulasjonsfrekvens = 50

bærebølge = sinus_generator(frekvens, v)
modulasjonsbølge = sinus_generator(modulasjonsfrekvens, v)
modulasjonsbølge = (modulasjonsbølge + 1) / 2            # "Forskyver" verdiene fra [-1,1] til [0,1]
resultat = bærebølge * modulasjonsbølge                  # Multipliser bølgene


spill_av(resultat)



# ---------------------------------------------------------
# Kode for å plotte grafene (ikke nødvendig å forstå denne)

antall_samples = 2400
fig, axs = subplots(3)

fig.suptitle("Amplitudemodulasjon")
fig.set_figwidth(15)
fig.set_figheight(6)

subplots_adjust(hspace=0.7)
x = arange(0, antall_samples)

axs[0].set_title("Bærebølge")
axs[1].set_title("Modulasjonsbølge")
axs[2].set_title("Resultat")
axs[0].plot(x, bærebølge[0:antall_samples])
axs[1].plot(x, modulasjonsbølge[0:antall_samples])
axs[2].plot(x, resultat[0:antall_samples])
[g.xaxis.set_visible(False) for g in axs]

show()
```

Den første bølgen er en enkel sinusbølge. Denne multipliseres med en annen sinusbølge som har en lavere frekvens og som beveger seg mellom 0 og 1. Resultatet er en bølge som varierer (moduleres) i lydstyrke! I musikk kalles denne effekten for *tremolo*.

Amplitudemodulasjon er en viktig teknikk innenfor flere områder. Når man lager digitale instrumenter eller analoge instrumenter/synther er amplitudemodulasjon én måte å lage lyder som er mer interessante enn kun enkle bølgeformer. Sjekk for eksempel ut denne demonstrasjonen: https://youtu.be/VHZ0KOvRj-c?t=35

Amplitudemodulasjon brukes også i radiokommunikasjon for å overføre signaler, for eksempel et lydsignal fra en radiostasjon. I det tilfellet er det lydsignalet man overfører som er modulasjonsbølgen. Lydsignalet "bæres" da av bærebølgen (engelsk: *carrier*). Mottakeren kan lese av den modulerte bærebølgen og gjøre det om igjen til et hørbart lydsignal. En annen vanlig modulasjonsteknikk i radiokommunikasjon er frekvensmodulasjon (FM). Før DAB ble introdusert brukte man frekvensmodulasjon til radiokringkasting i Norge.
